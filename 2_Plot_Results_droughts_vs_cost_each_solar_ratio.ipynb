```python
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
FACTORS_CHANGE_DUNKLEFLAUTE = ["03", "04", "06", "07", "08", "13", "15", "17", "20", "25", "30"]   # "03" for 0.3 times "more" frequent, "17" for 1.7 times more frequent
SIGNS = ["less", "less", "less", "less", "less", "more", "more", "more", "more", "more", "more"] # "less" for less frequent, "more" for more frequent
```

```python
FACTORS_CHANGE_DUNKLEFLAUTE = ["03","07","08", "13", "15", "17", "20", "25", "30"]   # "03" for 0.3 times "more" frequent, "17" for 1.7 times more frequent
SIGNS = ["less", "less","less", "more", "more", "more", "more", "more", "more"] # "less" for less frequent, "more" for more frequent
```

```python
factor_change_dunkleflaute="03"
sign="less"
```

```python
Obj_value_drought_Mass = pd.read_csv("synthetic"+factor_change_dunkleflaute+"x"+sign+".csv")
```

```python
plt.figure(figsize=(8,6))

scatter = plt.scatter(
    Obj_value_drought_Mass['objective_value'], 
    Obj_value_drought_Mass['drought_mass'], 
    c=Obj_value_drought_Mass['solar_ratio'], 
    cmap='viridis',
    s=50,
    edgecolor='k'
)

plt.colorbar(scatter, label='Solar ratio')
plt.xlabel('Objective value')
plt.ylabel('Drought mass')
plt.title('Objective value vs drought mass, with respect to solar ratio (' + str(int(factor_change_dunkleflaute) / 10) + ' Dunkleflaute)')
plt.grid(True)

# Find index of minimum objective_value
min_idx = Obj_value_drought_Mass['objective_value'].idxmin()

# Get coordinates of that point
x_min = Obj_value_drought_Mass.loc[min_idx, 'objective_value']
y_min = Obj_value_drought_Mass.loc[min_idx, 'drought_mass']
solar_ratio_min = Obj_value_drought_Mass.loc[min_idx, 'solar_ratio']

# Annotate with an arrow and text
plt.annotate(
    f'Solar ratio at min obj value: {solar_ratio_min:.2f}',  # Text with 2 decimals
    xy=(x_min, y_min),  # Point to annotate
    xytext=(x_min + 1e8, y_min + 0.05),  # Text position (offset a bit)
    arrowprops=dict(facecolor='black', arrowstyle='->'),
    fontsize=10,
    bbox=dict(boxstyle="round,pad=0.3", fc="yellow", alpha=0.5)
)

plt.show()
```

```python
for factor_change_dunkleflaute, sign in zip(FACTORS_CHANGE_DUNKLEFLAUTE, SIGNS):
    Obj_value_drought_Mass = pd.read_csv("synthetic"+factor_change_dunkleflaute+"x"+sign+".csv")
    plt.figure(figsize=(8,6))

    scatter = plt.scatter(
        Obj_value_drought_Mass['objective_value'], 
        Obj_value_drought_Mass['drought_mass'], 
        c=Obj_value_drought_Mass['solar_ratio'], 
        cmap='viridis',
        s=50,
        edgecolor='k'
    )

    plt.colorbar(scatter, label='Solar ratio')
    plt.xlabel('Objective value')
    plt.ylabel('Drought mass')
    plt.title('Objective value vs drought mass, with respect to solar ratio (' + str(int(factor_change_dunkleflaute) / 10) + ' Dunkleflaute)')
    plt.grid(True)
    
    # Find index of minimum objective_value
    min_idx = Obj_value_drought_Mass['objective_value'].idxmin()

    # Get coordinates of that point
    x_min = Obj_value_drought_Mass.loc[min_idx, 'objective_value']
    y_min = Obj_value_drought_Mass.loc[min_idx, 'drought_mass']
    solar_ratio_min = Obj_value_drought_Mass.loc[min_idx, 'solar_ratio']

    # Annotate with an arrow and text
    plt.annotate(
        f'Solar ratio at min obj value: {solar_ratio_min:.2f}',  # Text with 2 decimals
        xy=(x_min, y_min),  # Point to annotate
        xytext=(x_min + 1e8, y_min + 0.05),  # Text position (offset a bit)
        arrowprops=dict(facecolor='black', arrowstyle='->'),
        fontsize=10,
        bbox=dict(boxstyle="round,pad=0.3", fc="yellow", alpha=0.5)
    )
    plt.savefig("Objective_vs_DroughtMass_SolarRatio_Dunkleflaute"+factor_change_dunkleflaute+"x"+sign+".pdf")
    plt.show()
    
```
